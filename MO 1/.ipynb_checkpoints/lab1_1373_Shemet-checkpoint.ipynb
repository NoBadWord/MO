```python
mean_value = df['IQ'].mean()
print(f'Среднее значение: {mean_value}')
```

```python
std_value = df['IQ'].std()
print(f'Стандартное отклонение: {std_value}')
```

```python
values = df['IQ']
```

```python
plt.hist(values, bins=10, edgecolor='k')  # bins - количество столбцов гистограммы
plt.xlabel('Значение')  # подпись оси X
plt.ylabel('Частота')   # подпись оси Y
plt.title('Гистограмма распределения значений')  # название гистограммы
plt.show()
```

```python
df.hist (column='IQ')
```

```python
sns.boxplot(x=df['IQ'])
```

```python
plt.boxplot(values)
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df_cleaned = df.dropna()
```

```python
mis_values = df_cleaned.isnull().sum()
print(mis_values)
```

```python
correlation_matrix = df_cleaned.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Матрица корреляции')
plt.show()
```

```python
sns.pairplot(df_cleaned)
plt.show()
```
