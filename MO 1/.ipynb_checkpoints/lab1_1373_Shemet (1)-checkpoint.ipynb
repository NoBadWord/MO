Для каждого атрибута построим "ящики с усами", чтобы определить выбросы. Если значение находится вне "усов", то это - выброс. Используем функцию библиотеки seaborn.

```python
sns.boxplot(x=df['IQ'])
```

```python
sns.boxplot(x=df['education_expenditure'])
```

```python
sns.boxplot(x=df['avg_income'])
```

```python
sns.boxplot(x=df['avg_temp'])
```

```python
sns.boxplot(x=df['rank'])
```

Как мы видим, выбросы есть у всех атрибутов, кроме rank и avg_temp

Проверим, есть ли пропущенные значения. Для этого используем функцию библиотеки pandas, которая определяет количество пропущенных значений для каждого атрибута.

```python
empty_values = df.isnull().sum()
print(empty_values)
```

Пропущенные значения есть; их количество видно выше.
Вариант обработки пропущенных значений - удалить из датасета строки с пропущенными значениями, используя функцию библиотеки pandas.

```python
df_new = df.dropna()
```

Строки удалены, еще раз проверим, что пропущенных значений больше нет:

```python
empty_or_not_values = df_new.isnull().sum()
print(empty_or_not_values)
```

Для определения высокоррелированных атрибутов построим матрицу корреляции, а для ее визуализации - тепловую карту. На тепловой карте более яркие цвета обозначают высокоррелированные параметры, менее яркие - низкоколеррированные. Используем функции библиотек seaborn и matplotlib.

```python
sns.heatmap(df_new.corr(), annot=True, cmap='coolwarm',linewidths=0.5, linecolor='black' )
plt.title('Матрица корреляции')
plt.show()
```

Как видно из тепловой карты, высокая положительная корреляция у пар атрибутов: avg_income и education_expenditure.

Высокая отрицательная корреляция у пар атрибутов: IQ
 и rank

 Для этих выводов мы взяли, что высокая корреляция >0.7 или <-0.7

 Самая низкая корреляция у атрибутов avg_temp и avg_income. Можно сказать, что она отсутствует.

Построим матрицу графиков рассеивания. Используем функции библиотек seaborn и matplotlib.

Вывод:

Проанализируем полученные графики рассеивания.

На диагонали матрицы находятся гистограммы распределения,они могут помочь оценить форму распределения.

Графики, расположенные вне диагонали, представляют собой диаграммы рассеивания для каждой пары атрибутов. Они позволяют оценить, существует ли какая-то зависимость или взаимосвязь между параметрами атрибутов.

У многих диаграмм рассеивания есть необычные точки, выбивающиеся от остальных - это выбросы.

Если два атрибута сильно коррелированы, это может свидетельствовать о наличии линий в зависимости, что мы и наблюдаем у диаграммы рассеивания rank и IQ.
